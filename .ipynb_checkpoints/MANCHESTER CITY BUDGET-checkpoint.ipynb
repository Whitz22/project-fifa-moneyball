```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Viewing DATA
Fifa=pd.read_csv("Fifa 23 Players Data.csv")
Fifa.head(30)
```

```python
#Checking for the columns
Fifa.columns
```

```python
#Selecting relevant Columns
FIFA=Fifa[['Known As', 'Overall', 'Potential', 'Value(in Euro)',
       'Positions Played', 'Best Position', 'Age',
       'Height(in cm)', 'Weight(in kg)', 'Club Name', 'Wage(in Euro)', 
        'Release Clause', 'Club Position',
       'Contract Until', 'Joined On', 'On Loan']]
display(FIFA)
```

```python
#Calculating for gap potential
FIFA['Gap_Potential'] = FIFA['Potential'] - FIFA['Overall']

FIFA.head()
```

```python
#finding players less than Age 25 and gap potential above 10
FIFA['Under_25_and_GP_above_10']= (FIFA['Age']<25) & (FIFA["Gap_Potential"]>10)
display(FIFA)
```

```python
#Finding Nbr of MC players in the dataSet
Manchester_Players = FIFA['Club Name'].value_counts()
count_of_Manchester_Players = Manchester_Players['Manchester City']
print(count_of_Manchester_Players)
```

```python
#Finding players in Manchester city
FIFA[(FIFA['Club Name']=='Manchester City')].head(27)
```

```python
# Finding the total budget of MC players by calculating thier total Wages
club_name = 'Manchester City'

Manchester_wage = FIFA[FIFA['Club Name'] == club_name]

Total_wages_in_Manchester_city = Manchester_wage['Wage(in Euro)'].sum()

print(f"The total wages of players in {club_name} is: {Total_wages_in_Manchester_city}")
```

```python
# Find the total value of MC players by calculating thier Value(in Euros)
club_name = 'Manchester City'

Manchester_value = FIFA[FIFA['Club Name'] == club_name]

Total_value_in_Manchester_city = Manchester_value['Value(in Euro)'].sum()

print(f"The total Value(in Euros) of players in {club_name} is: {Total_value_in_Manchester_city}")
```

```python
#finding the difference in Value and Total Wages of MC players
Difference=Total_value_in_Manchester_city - Total_wages_in_Manchester_city
Difference
```

```python
#finding the average salary of the postions played in MC

club_name = 'Manchester City'

Manchester_players = FIFA[FIFA['Club Name'] == club_name]

Average_salary_per_position = Manchester_players.groupby(['Positions Played', 'Club Name'])['Wage(in Euro)'].mean().reset_index()

display(Average_salary_per_position)
```

```python
 #Plotting the average wages of MC players
sorted_wages = Average_salary_per_position.sort_values(by='Wage(in Euro)')
plt.figure(figsize=(12, 6))
plt.bar(sorted_wages['Positions Played'], sorted_wages['Wage(in Euro)'])
plt.xlabel('Positions Played')
plt.ylabel('Average Wage (in Euro)')
plt.title(f'Average Wage of Players in {club_name} by Position')
plt.xticks(rotation=45) 

plt.show()
```

```python
#Rank players by their Value within "MC" 
club_name = 'Manchester City'

rank_Manchester_players = FIFA[FIFA['Club Name'] == club_name]

ranked_manchester_players_value = rank_Manchester_players .sort_values(by='Value(in Euro)', ascending=False)

display(ranked_manchester_players_value)
```

```python
#Selceting players within MC to be transferred based on *old age and  low GP<10
Manchester_City=FIFA[(FIFA['Club Name'] == 'Manchester City') & (FIFA['Under_25_and_GP_above_10'] ==False)]
display(Manchester_City)
```

```python
"""Based on MC Data above, the Data Analysts in MC has asked management to transfer the players below based on age(above_25) and low Gap potential

1.Bernardo Silva
2.R. Mahrez
3.M. Akanji
4.S. Ortega
"""
```

```python
#Total Wages of selected players to be transfered

Wages = (Manchester_City['Known As'] == 'M.Akanji') | (Manchester_City['Known As'] == 'S.Ortega')|(Manchester_City['Known As'] == 'Bernado Silva')| (Manchester_City['Known As'] == 'R. Mahrez')


Transfer_players_wages = Manchester_City[Wages]


total_wage = Transfer_players_wages['Wage(in Euro)'].sum()

print(f"The total wages of selected players to be transfered is: {total_wage}")
```

```python
#Total Market Value of selected MC players to be transfered
Values = (Manchester_City['Known As'] == 'M.Akanji') | (Manchester_City['Known As'] == 'S.Ortega')|(Manchester_City['Known As'] == 'Bernado Silva')| (Manchester_City['Known As'] == 'R. Mahrez')


Transfer_players_value = Manchester_City[Values]


total_value = Transfer_players_value['Value(in Euro)'].sum()

print(f"The total Value(in Euros) of selected players to be transfered is: {total_value}")
```
