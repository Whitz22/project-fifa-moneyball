```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Viewing DATA
Fifa=pd.read_csv("Fifa 23 Players Data.csv")
Fifa.head(30)
```

```python
#Selecting relevant Columns
FIFA=Fifa[['Known As', 'Overall', 'Potential', 'Value(in Euro)',
       'Positions Played', 'Best Position', 'Age',
       'Height(in cm)', 'Weight(in kg)', 'Club Name', 'Wage(in Euro)', 
        'Release Clause', 'Club Position',
       'Contract Until', 'Joined On', 'On Loan']]
display(FIFA)
```

```python
FIFA['Gap_Potential'] = FIFA['Potential'] - FIFA['Overall']

FIFA.head()
```

```python
#finding players less than Age 25 and gap potential above 10
FIFA['Under_25_and_GP_above_10']= (FIFA['Age']<25) & (FIFA["Gap_Potential"]>10)
display(FIFA)
```

```python
#Filtering for  Potential GK in our FIFA with these conditions

FIFA[(FIFA['Positions Played'] == 'GK') & (FIFA['Under_25_and_GP_above_10'] == True) & (FIFA['Overall']>65) | (FIFA['Contract Until']==2024)]
```

```python
#Filtering for  Potential GK in our FIFA with these conditions
#I was thinking to invest for an older GK with many years experience

FIFA[(FIFA['Positions Played'] == 'GK') & (FIFA['Overall']>85)]
```

```python
#Filtering for  Potential CM,CAM in our FIFA with these conditions
FIFA[(FIFA['Positions Played'] == 'CAM,CM') & (FIFA['Under_25_and_GP_above_10'] == True) & (FIFA['Overall']>68) | (FIFA['Contract Until']==2023)]
```

```python
#Filtering for  Potential CB in our FIFA with these conditions

FIFA[(FIFA['Positions Played'] == 'CB') & (FIFA['Under_25_and_GP_above_10'] == True) & (FIFA['Overall']>70)] #| (Sample_Fifa['Contract Until']==2023)]
```
