```python
# Cleaning the data by selecting the columns we will use for our analysis.

Fifa_Cle = Fifa[['Known As', 'Overall', 'Potential', 'Value(in Euro)',
       'Positions Played', 'Best Position', 'Age',
       'Height(in cm)', 'Weight(in kg)',
       'Club Name', 'Wage(in Euro)', 'Release Clause', 'Club Position',
       'Contract Until', 'Joined On', 'On Loan', 'National Team Name','ST Rating',
       'LW Rating', 'LF Rating', 'CF Rating', 'RF Rating', 'RW Rating',
       'CAM Rating', 'LM Rating', 'CM Rating', 'RM Rating', 'LWB Rating',
       'CDM Rating', 'RWB Rating', 'LB Rating', 'CB Rating', 'RB Rating',
       'GK Rating','Nationality']]
```

```python
# Add a column to understand the potential gap of each player.

Fifa_Cle["Gap_Potential"] = Fifa_Cle["Potential"] - Fifa_Cle["Overall"]
Fifa_Cle
```

```python
#Get the data of all the players at Manchester City to have an overview of the players scores and potentials.

Fifa_MC = Fifa_Cle.loc[Fifa_Cle['Club Name'] == "Manchester City"]
Fifa_MC
```

```python
# Get the data of all the players outside of Manchester City.

Fifa_Other = Fifa_Cle.loc[Fifa_Cle['Club Name'] != "Manchester City"]
Fifa_Other
```

***ANALYSIS 1 - Determining players to transfer according to their overall and potential. ***

```python
# Create a new DataFrame to focus on the ratings of all the players (in and out of MC)

Fifa_Other.columns
```

```python
Fifa_MP = Fifa_Cle[['Known As', 'Overall', 'Potential','Positions Played','Club Name',
       'Club Position', 'ST Rating', 'LW Rating', 'LF Rating',
       'CF Rating', 'RF Rating', 'RW Rating', 'CAM Rating', 'LM Rating',
       'CM Rating', 'RM Rating', 'LWB Rating', 'CDM Rating', 'RWB Rating',
       'LB Rating', 'CB Rating', 'RB Rating', 'GK Rating', 'Nationality',
       'Gap_Potential']]
```

```python
# Add the mean value of the "Overall" column for each positions.
#With this code, we could get all the means of each rating per position.
#Estimating that most likely the overall score was already calculated according to the other ratings, we decided to focus on that column.

Fifa_MP.groupby(['Positions Played']).mean()

#Fifa_MP["Overall_Mean"] = Fifa_MP.groupby(['Club Position']).agg({"Overall":"mean"})
#Fifa_MP
```

```python
#Calculate the mean values for each position:
position_means = Fifa_MP.groupby(['Positions Played'])['Overall'].mean().reset_index()

#Rename the "Overall" column with the position_means DataFrame to "Mean Overall".
position_means.rename(columns={'Overall': 'Mean Overall'}, inplace=True)

#Merge the mean values back into your initial DataFrame based on the "Club Position" column:
Fifa_MP = Fifa_MP.merge(position_means, on='Positions Played', how='left')

Fifa_MP
```

```python
#Re-order the columns to have "Overall" next to "Mean Overall"

Fifa_MP.columns
```

```python
Fifa_MP = Fifa_MP[['Known As','Overall', 'Mean Overall','Potential','Positions Played','Club Position', 'Club Name', 'ST Rating',
       'LW Rating', 'LF Rating', 'CF Rating', 'RF Rating', 'RW Rating',
       'CAM Rating', 'LM Rating', 'CM Rating', 'RM Rating', 'LWB Rating',
       'CDM Rating', 'RWB Rating', 'LB Rating', 'CB Rating', 'RB Rating',
       'GK Rating', 'Nationality', 'Gap_Potential']]

Fifa_MP
```

```python
Fifa_MC = Fifa_MP.loc[Fifa_MP['Club Name'] == "Manchester City"]
Fifa_MC
```

```python
"""

This is where we realized that we only have substitutes and reserve club positions players with an overall score below the mean score of all the other players.
We took the decision to instead take the mean of the Top 10 champions league teams of 2023-2024 which are the following:

1. Manchester City
2. Real Madrid
3. FC Barcelona
4. Bayern Munich
5. Inter
6. Real Sociedad
7. Arsenal
8. PSG
9. Atletico de Madrid
10. Porto

"""
```

```python
#Creating a new DataFrame with the data of the top 10 teams in our Champions league ranking.

Fifa_Champions = Fifa[Fifa['Club Name'].isin(["Manchester City", "Real Madrid", "FC Barcelona","Bayern Munich","Inter","Real Sociedad", "Arsenal","Paris Saint-Germain", "Atletico de Madrid","Porto"])]

Fifa_Champions
```

```python
#Calculate the mean values for each position:
position_means2 = Fifa_Champions.groupby(['Positions Played'])['Overall'].mean().reset_index()

#Rename the "Overall" column with the position_means DataFrame to "Mean Overall".
position_means2.rename(columns={'Overall': 'Mean Overall'}, inplace=True)

#Merge the mean values back into your initial DataFrame based on the "Club Position" column:
Fifa_Champions = Fifa_Champions.merge(position_means2, on='Positions Played', how='left')

Fifa_Champions

```

```python
Fifa_Champions.columns
```

```python
Fifa_Champions = Fifa_Champions[['Known As', 'Overall','Mean Overall', 'Potential','Positions Played','Club Name', 'Value(in Euro)','Age', 'Best Position', 'Wage(in Euro)', 'Release Clause',
       'Club Position', 'Contract Until', 'Joined On', 'On Loan',
       'National Team Name', 'ST Rating', 'LW Rating', 'LF Rating',
       'CF Rating', 'RF Rating', 'RW Rating', 'CAM Rating', 'LM Rating',
       'CM Rating', 'RM Rating', 'LWB Rating', 'CDM Rating', 'RWB Rating',
       'LB Rating', 'CB Rating', 'RB Rating', 'GK Rating', 'Nationality']]

Fifa_Champions
```

```python
#Checking again the overall score of our players in Manchester City compared to the mean of the Champions League top teams.

Fifa_MC2 = Fifa_Champions.loc[Fifa_Champions['Club Name'] == "Manchester City"]
Fifa_MC2
```

```python
#As we also wanted to take into account the potential, we created a new column taking it inot account.

Fifa_MC2["Real_OScore"] = (Fifa_MC2["Overall"] + Fifa_MC2["Potential"])/ 2

Fifa_MC2
```

```python
#Focusing our data on the columns needed to determine who we would like to transfer:

Fifa_MC2.columns
```

```python
Fifa_MC2 = Fifa_MC2[['Known As', 'Overall','Potential', 'Real_OScore','Mean Overall', 'Club Name',
       'Value(in Euro)','Positions Played', 'Club Position']]

Fifa_MC2
```

```python
#Creating a new column with a boolean analysis to highlight players with a Real_OScore below the mean of the Champions leage top 10 teams.

Fifa_MC2["A1_Transfer?"] = np.where(Fifa_MC2["Real_OScore"] < Fifa_MC2["Mean Overall"], True, False)
Fifa_MC2
```

***ANALYSIS 2 - Determining players to transfer according to their age and release date. ***

```python
Fifa_Cle.columns

```

```python
Fifa_A2 = Fifa_MC2

Fifa_A2["Age"] = Fifa_Champions["Age"]
Fifa_A2["Wage (in euro)"] = Fifa_Champions["Wage(in Euro)"]
Fifa_A2["Release Clause"] = Fifa_Champions["Release Clause"]
Fifa_A2["Contract Until"] = Fifa_Champions["Contract Until"]

Fifa_A2
```

```python
Fifa_CU = Fifa_A2[Fifa_A2['Contract Until'].isin(["2023","2024","2025"])]
Fifa_CU


```

```python
age_threshold = 25
oscore_threshold = 5  # determine the difference from the mean that we will target.
mean_overall = Fifa_CU['Mean Overall']  # Store the mean overall in a variable for efficiency.

highlight_mask = (
    (Fifa_CU['Age'] > age_threshold) &  # Age condition
    (
        (abs(Fifa_CU['Real_OScore'] - mean_overall) <= oscore_threshold) |  # Real_OScore condition
        (Fifa_CU['Real_OScore'] < mean_overall)  # Include players with Real_OScore below the mean
    )
)

# Create a new column "A2_Transfer?" with True for the players we will transfer, and False otherwise.
Fifa_CU['A2_Transfer?'] = highlight_mask

Fifa_CU
```
